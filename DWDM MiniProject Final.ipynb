```python
# importing necessary libraries
import pandas as pd
import numpy as np
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("Music_dataset.csv")
```

```python
#scatter plot for the given data without KMode
plt.scatter(data['index'],data['genre'])
plt.xlabel("Music")
plt.ylabel("Genre")
plt.title("ScatterPlot based on music Genres")
```

```python
# Building the model with 7 clusters
kmode = KModes(n_clusters=7, init = "random", n_init = 50)
clusters = kmode.fit_predict(data)
clusters
```

```python
#data=data.drop('Cluster',1)
data.insert(10, "Cluster", clusters, True)
data
```
